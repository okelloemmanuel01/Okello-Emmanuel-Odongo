```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the csv file
df = pd.read_csv('csv_files/data.csv')
```

```python
#check the sum of null values
df.isnull().sum()
```

```python
#Fill the empty ones with the mean
df['Calories'] = df['Calories'].fillna(df['Calories'].mean())
```

```python
df.isnull().sum()
```

```python
#check for duplicates 
print(df.duplicated().sum())
```

```python
#Remove duplicates 
df = df.drop_duplicates()
```

```python
# Set up the figure (1 row, 2 columns)
plt.figure(figsize=(15, 5))  # Width=15", Height=5"

# --- Plot 1: Histogram (Duration Distribution) ---
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
sns.histplot(data=df, x='Duration', kde=True, bins=20, color='skyblue')
plt.title('Distribution of Duration')
plt.xlabel('Duration (mins)')
plt.ylabel('Count')

# --- Plot 2: Bar Plot (Average Pulse by Duration) ---
plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
sns.barplot(data=df, x='Duration', y='Pulse', estimator=np.mean, ci=None, palette='viridis')
plt.title('Average Pulse by Duration')
plt.xlabel('Duration (mins)')
plt.ylabel('Mean Pulse (bpm)')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Set up the figure (1 row, 2 columns)
plt.figure(figsize=(15, 5))  # Width=15", Height=5"

# --- Plot 1: Histogram (Maxpulse Distribution) ---
plt.subplot(1, 2, 1)  # Left plot
sns.histplot(data=df, x='Maxpulse', kde=True, bins=20, color='salmon')
plt.title('Distribution of Maxpulse')
plt.xlabel('Maxpulse (bpm)')
plt.ylabel('Count')

# --- Plot 2: Bar Plot (Average Calories by Duration) ---
plt.subplot(1, 2, 2)  # Right plot
sns.barplot(data=df, x='Duration', y='Calories', estimator=np.mean, ci=None, palette='rocket')
plt.title('Average Calories by Duration')
plt.xlabel('Duration (mins)')
plt.ylabel('Mean Calories Burned')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Calculate correlations between numerical columns (including Calories)
corr_matrix = df[['Duration', 'Pulse', 'Maxpulse', 'Calories']].corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, 
            annot=True,  # Show correlation values
            cmap='coolwarm',  # Blue-to-red color scale
            vmin=-1, vmax=1,  # Fix scale from -1 to 1
            linewidths=0.5,
            fmt=".2f")  # 2 decimal places

plt.title('Correlation Heatmap: Calories vs Other Metrics', pad=20)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
